```python
from iscatter import *
from pandas.io import wb
```

```python
df4 = wb.download(indicator=['SP.DYN.TFRT.IN','SP.DYN.LE00.IN'], country=['CN', 'GB', 'BD'], start=1975, end=2005)
df4=df4.unstack('country')
df4.columns=[' '.join(col).strip() for col in df4.columns.values]
df4.reset_index(inplace=True)
df4
```

```python
name='Fertility rate and Life expectancy at birth'
desc='Fertility rate (SP.DYN.TFRT.IN) and life expectancy at birth(SP.DYN.LE00.IN) for several countries by year. Source: World Bank Development Indicator '
xinit='year'
yinit='SP.DYN.TFRT.IN Bangladesh'
ainit=''
xattr=[]#['income']
yattr=['year']#['index']
colour=''
STUB='fertlife'

#Note: I extended the minified version of the iScatter js to accept axis range setting
iScatterGen(df4,STUB,name,desc,xinit,yinit,ainit,xattr=xattr,yattr=yattr,yrange=[0,8])#yrange=[0,8]
```

```python
df5 = wb.download(indicator=['SP.DYN.TFRT.IN','SP.DYN.LE00.IN','SH.DYN.MORT'], country=['CN', 'GB', 'BD'], start=1975, end=2005)
df5=df5.unstack('country')
df5.columns=[' '.join(col).strip() for col in df5.columns.values]
df5.reset_index(inplace=True)
df5[:3]
```

```python
name='Fertility rate, Life expectancy and Under 5 Mortality at birth'
desc='''Fertility rate (SP.DYN.TFRT.IN), life expectancy at birth(SP.DYN.LE00.IN) and
under 5 mortality (SH.DYN.MORT) for several countries by year. Source: World Bank Development Indicator '''
xinit='SH.DYN.MORT China'
yinit='SP.DYN.LE00.IN China'
ainit='SP.DYN.TFRT.IN China'
xattr=[]#['income']
yattr=['year']#['index']
colour=''
STUB='fertlife5mort'

#Note: I extended the minified version of the iScatter js to accept axis range setting
iScatterGen(df5,STUB,name,desc,xinit,yinit,ainit,xattr=xattr,yattr=yattr)#yrange=[0,8]
```
